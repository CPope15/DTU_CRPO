# Copy master files

Uses the read_hawc2 function within Hawc dictionary class to read HAWC2 inputs into python dictionaries and then write them out into this current directory.
In this case i am making copies of the manually translated model from Github.
The goal would be first get the master file here with all needed supplementing files. It could run from this location now but it is now fully in python so all files can rapidly be re-written and new runs & data produced easily

This is not purely the GITHub model - changes have been made such as the detuned controller parameters.

```python
#why do i choose this as the master file? Doesn't really matter.
master_file = "IEA_15MW_RWT_UMaineSemi"

iea15_manual_repo_H2 = HAWC2_dict().read_hawc2(
    iea15_manual_repo_fpath, "htc\%s.htc"%master_file)

iea15_temp = iea15_manual_repo_H2.copy()
```

```python
iea15_temp['htc_filename']='htc/%s_template.htc'%master_file
iea15_temp["htc"]["simulation"]["visualization"] = "visualization/%s.hdf5"%master_file
iea15_temp["htc"]["simulation"]["logfile"] = "log/%s.log"%master_file
iea15_temp["htc"]["new_htc_structure"]["ext_sys"][0]["data"][0][1] = './HydroData/IEA-15-240-RWT-UMaineSemi'
```

```python
iea15_temp.write_hawc2(".")
```
